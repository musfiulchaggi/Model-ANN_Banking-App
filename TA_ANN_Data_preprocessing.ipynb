```python
import tensorflow 
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
# initializing the ANN model
classifier=Sequential()
```

```python
# classifier.add(Dense(6, activation = 'relu',input_dim=10))
classifier.add(Dense(6, activation = 'relu',input_dim=5))
classifier.add(Dense(6,activation='relu'))
classifier.add(Dense(1,activation='sigmoid'))
```

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
classifier.summary()
```

```python
ACCURACY_THRESHOLD = 0.99
```

```python
# Implement callback function to stop training
# when accuracy reaches e.g. ACCURACY_THRESHOLD = 0.95
class myCallback(keras.callbacks.Callback): 
    def on_epoch_end(self, epoch, logs={}): 
        if(logs.get('accuracy') > ACCURACY_THRESHOLD):   
            print("\nReached %2.2f%% accuracy, so stopping training!!" %(ACCURACY_THRESHOLD*100))   
            self.model.stop_training = True
```

```python
callbacks = myCallback()
```

```python
# batch_size = 1
# epochs = 200
# shuffle = False


history = classifier.fit(X_train,y_train,batch_size=10,epochs=100,verbose=1,callbacks = [callbacks],validation_split=0.25)
```

```python
# list all data in history

print(history.history.keys())
```

```python
# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
